```python
import pandas as pd
import re
import konlpy
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
import numpy as np
from tqdm import tqdm
import urllib.request
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('data/dcinside.xlsx')
```

```python
okt = konlpy.tag.Okt()

def text_preprocess(x):
    text=[]
    a = re.sub('[^가-힣0-9a-zA-Z\\s]', '',x)
    for j in a.split():
        text.append(j)
    return ' '.join(text)

def tokenize(x):
    text = []
    tokens = okt.pos(x)
    for token in tokens :
        #if token[1] == 'Adjective' or token[1]=='Adverb' or token[1] == 'Determiner' or token[1] == 'Noun' or token[1] == 'Verb' or 'Unknown':
        if token[1] == 'Noun' :    
            text.append(token[0])
    return text
```

```python
def get(x):
    result=""
    for i in x:
        result=result+" "+i
    return result
```

```python
df['token'] = df['제목'].apply(lambda x : text_preprocess(x))
df['token'] = df['token'].apply(lambda x: tokenize(x))
df['data']=df['token'].apply(lambda x: get(x))
```

```python
N=len(df)
from math import log
def tf(t, d):
    return d.count(t)
def idf(t):
    d=0
    for doc in df['data']:
        d += t in doc
    return d

def tfidf(t, d):
    return tf(t, d)*idf(t)
```

```python
result=[]
for i in range(N):
    result.append([])
    d=df.loc[i,'data']
    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tf(t,d))
tf_ = pd.DataFrame(result, columns = vocab)
tf_
```
