```python
pages = driver.find_elements_by_css_selector('a.board-nav-page')
df = pd.DataFrame()
# for j in range(len(pages)) : # 페이지 넘기기
for j in range(2) : # 페이지 넘기기(2페이지만)
    driver.get(f'https://www.clien.net/service/board/park?&od=T31&category=0&po={j}')
    if j == 0 : # 1페이지의 경우 첫번째 게시물 : 공지 / 두번째 게시물 : 광고

        # 게시물
        for i in tqdm(range(30)) : 
            info_dic = {}
            heads = driver.find_elements_by_css_selector('a.list_subject') # 게시물 목록
            heads[i+2].click() # 게시물 링크 클릭
            time.sleep(2)
            
            try : 
                title = driver.find_element_by_css_selector('h3.post_subject > span').text # 제목
            except : 
                title = np.nan

            try : 
                num_likes = driver.find_element_by_xpath('//*[@id="div_content"]/div[2]/div/span').text # 공감수
            except : 
                num_likes = np.nan

            try :
                num_comments = driver.find_element_by_xpath('//*[@id="comment-point"]/a/strong').text # 댓글수
            except : 
                num_comments = np.nan

            try :
                user_name = driver.find_element_by_xpath('//*[@id="div_content"]/div[3]/div[1]/span/span/span').text # 글쓴이
            except : 
                user_name = np.nan

            try : 
                views = driver.find_element_by_xpath('//*[@id="div_content"]/div[3]/div[2]/span[1]/strong').text # 조회수
            except : 
                views = np.nan

            time.sleep(0.5)

            info_dic['제목'] = title; info_dic['공감수'] = num_likes; info_dic['댓글수'] = num_comments
            info_dic['글쓴이'] = user_name; info_dic['조회수'] = views

            df = df.append(info_dic, ignore_index = True)

            time.sleep(0.5)

            driver.get(f'https://www.clien.net/service/board/park?&od=T31&category=0&po={j}')
            time.sleep(2)

        print(df)
        
    else : 
        
        # 게시물
        for i in tqdm(range(30)) : 
            info_dic = {}
            heads = driver.find_elements_by_css_selector('a.list_subject') # 게시물 목록
            heads[i+1].click() # 게시물 링크 클릭
            time.sleep(2)
            
            try : 
                title = driver.find_element_by_css_selector('h3.post_subject > span').text # 제목
            except : 
                title = np.nan

            try : 
                num_likes = driver.find_element_by_xpath('//*[@id="div_content"]/div[2]/div/span').text # 공감수
            except : 
                num_likes = np.nan

            try :
                num_comments = driver.find_element_by_xpath('//*[@id="comment-point"]/a/strong').text # 댓글수
            except : 
                num_comments = np.nan

            try :
                user_name = driver.find_element_by_xpath('//*[@id="div_content"]/div[3]/div[1]/span/span/span').text # 글쓴이
            except : 
                user_name = np.nan

            try : 
                views = driver.find_element_by_xpath('//*[@id="div_content"]/div[3]/div[2]/span[1]/strong').text # 조회수
            except : 
                views = np.nan

            time.sleep(0.5)

            info_dic['제목'] = title; info_dic['공감수'] = num_likes; info_dic['댓글수'] = num_comments
            info_dic['글쓴이'] = user_name; info_dic['조회수'] = views

            df = df.append(info_dic, ignore_index = True)

            time.sleep(0.5)

            driver.get(f'https://www.clien.net/service/board/park?&od=T31&category=0&po={j}')
            time.sleep(2)

        print(df)
```
